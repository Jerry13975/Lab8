```python
# YOUR CODE HERE
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from scipy.stats.mstats import winsorize
from sklearn.linear_model import Ridge
```

```python
# YOUR CODE HERE
features = ['Log GDP per capita', 'Social support', 'Healthy life expectancy at birth',
            'Freedom to make life choices', 'Generosity', 'Perceptions of corruption',
            'Confidence in national government', 'Democratic Quality', 'Delivery Quality']
target = 'Life Ladder'
data = df[features + [target, 'year']].copy()

train = data[data['year'] < 2015]
val = data[data['year'] == 2016]
test = data[data['year'] == 2017]

X_train = train[features]
y_train = train[target]

X_val = val[features]
y_val = val[target]

X_test = test[features]
y_test = test[target]
```

```python
X_train_manual = X_train.copy()
X_val_manual = X_val.copy()
X_test_manual = X_test.copy()

means = X_train_manual.mean()
stds = X_train_manual.std()

X_train_manual = X_train_manual.fillna(means)
X_val_manual = X_val_manual.fillna(means)
X_test_manual = X_test_manual.fillna(means)

X_train_scaled = (X_train_manual - means)/stds
X_val_scaled = (X_val_manual - means)/stds
X_test_scaled = (X_test_manual - means)/stds

X_train_winsorized = X_train_scaled.copy()
X_val_winsorized = X_val_scaled.copy()
X_test_winsorized = X_test_scaled.copy()

for col in features: 
    limits = (.01, .01)
    X_train_winsorized[col] = winsorize(X_train_scaled[col], limits=limits)
    X_val_winsorized[col] = winsorize(X_val_scaled[col], limits=limits)
    X_test_winsorized[col] = winsorize(X_test_scaled[col], limits=limits)
```

```python
model = LinearRegression()
model.fit(X_train_winsorized, y_train)

y_val_pred = model.predict(X_val_winsorized)

r2 = r2_score(y_val, y_val_pred)
rmse = mean_squared_error(y_val, y_val_pred, squared=False)
print(f'Validation R^2:', r2)
print(f'Validation RMSE:', rmse)
```

```python
params = {'alpha': [.01, .1, 1, 10, 100]}
grid = GridSearchCV(Ridge(), param_grid=params, cv=5)
grid.fit(X_train_winsorized, y_train)

print('Best alpha:' , grid.best_params_['alpha'])

best_model = grid.best_estimator_
best_y_val_pred = best_model.predict(X_val_winsorized)

r2_best = r2_score(y_val, best_y_val_pred)
rmse_best = mean_squared_error(y_val, best_y_val_pred, squared=False)
print(f'Tuned Validation R^2: ', r2_best)
print(f'Tuned Validation RMSE: ', rmse_best)
```

```python
y_test_pred = best_model.predict(X_test_winsorized)

r2_test = r2_score(y_test, y_test_pred)
rmse_test = mean_squared_error(y_test, y_test_pred, squared=False)

print(f'Test R^2: ', r2_test)
print(f'Test RMSE: ', rmse_test)
```

```python
plt.scatter(y_test, y_test_pred)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.xlabel('Actual Life Ladder')
plt.ylabel('Predicted Life Ladder')
plt.show()
```

Since Linear Regression has no hyperparameters to tune, I chose to use Ridge Regression, which adds L2 regularization to prevent overfitting by penalizing large coefficients. However, even after tuning and selecting the best alpha, the R² and RMSE on the validation set did not improve significantly. This suggests that the model may be reaching the limit of what the current features can explain, or that the target variable contains a high level of noise, making it inherently difficult to predict.
